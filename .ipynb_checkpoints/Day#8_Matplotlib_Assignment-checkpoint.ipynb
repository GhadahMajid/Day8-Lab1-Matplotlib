```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
# List
plt.plot([1,2,0,4,5]);
```

```python
# Dataframe
plt.plot(pd.DataFrame([6,7,3,9]));
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie([2,6,9,10], explode=[.1,.1,.1,.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y, marker = 'o', s = 50, alpha = 0.4);
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.title('Scatter Plot', fontsize = 30, loc='left', color='r');
plt.xlabel('xlabel', fontsize=16)
plt.ylabel('ylabel', fontsize=16)
plt.scatter(x,y, marker = 'o', s = 50, alpha = 0.4);
```

## Q5: Create three line charts with Legends at left and colors

```python

plt.plot([1,2,0,4,5])
plt.plot([10,9,8,6,3])
plt.plot([8,5,8,2,0])
plt.legend(['First','Second','Third'], loc = 'upper left');
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])
plt.suptitle('Subplots with 3 charts',fontsize = 16)

#first
plt.subplot(1,3,1)
plt.pie(np.arange(1,5),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('First')

#second
plt.subplot(1,3,2) 
plt.pie(np.arange(6,10),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Second')

#third
plt.subplot(1,3,3) 
plt.pie(np.arange(11,15),labels=['Jan','Feb','Mar', 'Apr']);
plt.title('Third');

```
